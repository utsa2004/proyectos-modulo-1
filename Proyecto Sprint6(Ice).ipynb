```python
# Eliminar las 2 filas con valores ausentes en la columna "name", por no ser representativos
games = games.dropna(subset=["name"])
games["name"].isna().sum()
```

```python
# Eliminar las 2 filas con valores ausentes en la columna "genre", por no ser representativos
ames = games.dropna(subset=["genre"])
games["genre"].isna().sum()
```

```python
# Convertir la columna "year_of_release" de tipo "float" a tipo "int"
games["year_of_release"] = pd.to_numeric(games["year_of_release"], errors="coerce")
games["year_of_release"].unique()
```

```python
# Reemplazar la abreviatura "tbd" por NaN en la columna "user_score"
games["user_score"] = games["user_score"].replace("tbd", np.nan)
games["user_score"].unique()
```

```python
# Convertir la columna "user_score" de tipo "object" a tipo "float"
games["user_score"] = pd.to_numeric(games["user_score"], errors="coerce")
games["user_score"].dtype
```

```python
# Rellenar las filas con valores ausentes de la columna "rating", con la letra "U" que significa "Unknown"
games["rating"].fillna("U", inplace=True)
games["rating"].isna().sum()
```

```python
# Nueva columna con el significado de las letras de la columna "rating"

# Diccionario de mapeo
rating_map = {
    'E': 'Everyone',
    'E10+': 'Everyone 10 and older',
    'T': 'Teen',
    'M': 'Mature',
    'AO': 'Adults Only',
    'K-A': 'Kids to Adults',  # aunque es antiguo, fue reemplazado por "E" y ya no se usa oficialmente.
    'EC': 'Early Childhood',
    'U': 'Unknown',
    'RP': 'Rating Pending'
}

# Crear la nueva columna
games['rating_names'] = games['rating'].map(rating_map)
```

```python
# Ventas Totales Globales por A√±o de las Principales 10 Plataformas

# Obtener las 10 plataformas con mayores ventas totales
ventas_por_plataforma = games.groupby("platform")["total_sales"].sum().reset_index().sort_values(by="total_sales", ascending=False).reset_index(drop=True).head(10)

# Seleccionar las top 10 plataformas en ventas
top_10_plataformas = ventas_por_plataforma["platform"]

# Filtrar los datos originales para solo esas plataformas
datos_top_10_plataformas = games[games["platform"].isin(top_10_plataformas)]

# Agrupar por a√±o y plataforma para ver la evoluci√≥n de ventas
ventas_por_a√±o_y_plataforma = datos_top_10_plataformas.groupby(["year_of_release", "platform"])["total_sales"].sum().reset_index()

# Imorimir los resultados por a√±os de las 10 plataformas con mayores ventas
print(ventas_por_a√±o_y_plataforma)

# Crear el gr√°fico de l√≠neas
plt.figure(figsize=(10,5))
sns.lineplot(
    data=ventas_por_a√±o_y_plataforma,
    x="year_of_release",
    y="total_sales",
    hue="platform",
    marker="o"
)
plt.title("Ventas Totales Globales por A√±o de las 10 Plataformas Principales")
plt.xlabel("A√±o")
plt.ylabel("Ventas Totales Globales (Millones en USD)")
plt.legend(title='Plataforma')
plt.grid(False)

plt.show()
```

```python
# Revisi√≥n de los Datos de Ventas Totales Globales de las 10 Plataformas Principales
tabla_ventas_plataformas = ventas_por_a√±o_y_plataforma.pivot_table(
    index='platform',
    columns='year_of_release',
    values='total_sales',
    margins=True,
    margins_name='Total'
)
tabla_ventas_plataformas
```

```python
# Ventas Totales Globales de las Top 10 Plataformas por A√±os
ventas_plataformas_a√±os = tabla_ventas_plataformas.sum(axis=0).reset_index()
ventas_plataformas_a√±os.columns = ["year_of_release", 'total_sales']
ventas_plataformas_a√±os.head(26)
```

```python
# Determinaci√≥n del Per√≠odo de Vida de una Plataforma en la Industria de los Juegos a partir del a√±o 2000

# Filtrar plataformas que tengan al menos un lanzamiento en el rango 2000-2016
filtered_games = games[
    (games['year_of_release'] >= 2000) & (games['year_of_release'] <= 2016)
]

# Agrupar por plataforma en los datos filtrados
grouped_filtered = filtered_games.groupby('platform')

# Crear lista para resultados
results_filtered = []

for platform, group in grouped_filtered:
    first_year = group['year_of_release'].min()
    last_year = group['year_of_release'].max()
    duration = last_year - first_year
    results_filtered.append({
        'platform': platform,
        'first_year': first_year,
        'last_year': last_year,
        'duration': duration
    })

# Convertir a DataFrame
duration_filtered_games = pd.DataFrame(results_filtered)

# Calcular el promedio de duraci√≥n en este rango filtrado
promedio_duracion_filtrado = duration_filtered_games['duration'].mean()

print(duration_filtered_games)
print(f"EL Promedio del Per√≠odo de Vida de las Plataformas en el rango de a√±os del 2000 al 2016 es de: {promedio_duracion_filtrado:.2f} a√±os")
```

```python
# Ventas Globales Totales de las plataformas con ventas del lugar 11 al 20

# Obtener las plataformas ordenadas por ventas totales
ventas_por_plataforma = games.groupby("platform")["total_sales"].sum().reset_index().sort_values(by="total_sales", ascending=False).reset_index(drop=True)

# Seleccionar las plataformas del lugar 11 al 20 (√≠ndices 10 a 19)
plataformas_11_a_20 = ventas_por_plataforma.iloc[10:20]["platform"]

# Filtrar los datos originales para esas plataformas
datos_plataformas_11_a_20 = games[games["platform"].isin(plataformas_11_a_20)]

# Agrupar por a√±o y plataforma para ver la evoluci√≥n de ventas
ventas_por_a√±o_y_plataforma_11_a_20 = datos_plataformas_11_a_20.groupby(["year_of_release", "platform"])["total_sales"].sum().reset_index()

# Imorimir los resultados por a√±os de las 10 plataformas con mayores ventas
print(ventas_por_a√±o_y_plataforma_11_a_20)

# Crear el gr√°fico de l√≠neas
plt.figure(figsize=(12,8))
sns.lineplot(
    data=ventas_por_a√±o_y_plataforma_11_a_20,
    x="year_of_release",
    y="total_sales",
    hue="platform",
    marker="o"
)
plt.title("Ventas Globales Totales de Plataformas del Lugar 11 al 20")
plt.xlabel("A√±o")
plt.ylabel("Ventas Globales Totales (Millones en USD)")
plt.legend(title='Plataforma')
plt.grid(False)

plt.show()
```

```python
# C√°lculo de la Tasa de Crecimiento Anual(CAGR) de Plataformas

# Definir la Funci√≥n
def calculate_cagr(start_value, end_value, periods):
    return (end_value / start_value) ** (1 / periods) - 1

# Agrupar por plataforma y obtener ventas totales por a√±o
ventas_por_plataforma_a√±o = games.groupby(["platform", "year_of_release"])["total_sales"].sum().reset_index()

# Para cada plataforma, calcular CAGR entre el primer y √∫ltimo a√±o
cagr_list = []

for platform in ventas_por_plataforma_a√±o['platform'].unique():
    datos = ventas_por_plataforma_a√±o[ventas_por_plataforma_a√±o['platform'] == platform]
    start_year = datos['year_of_release'].min()
    end_year = datos['year_of_release'].max()
    start_sales = datos[datos['year_of_release'] == start_year]['total_sales'].values[0]
    end_sales = datos[datos['year_of_release'] == end_year]['total_sales'].values[0]
    periods = end_year - start_year + 1
    cagr = calculate_cagr(start_sales, end_sales, periods)
    cagr_list.append({'platform': platform, 'CAGR': cagr})

# Convertir a DataFrame y filtrar plataformas con CAGR positivo
cagr_df = pd.DataFrame(cagr_list)
potenciales_plataformas = cagr_df[cagr_df['CAGR'] > 0].sort_values(by='CAGR', ascending=False)

print(potenciales_plataformas)
```

**FORMULACION DE LAS HIPOTESIS RESPECTO A LAS CALIFICACIONES PROMEDIO DE LOS USUARIOS PARA LAS PLATAFORMAS XBOX ONE Y PC**

El objetivo del an√°lisis es determinar si existen diferencias significativas en las calificaciones promedio otorgadas por los usuarios a las plataformas Xbox One y PC. Se selecciona una prueba estad√≠stica bilateral (de dos colas) basada en la comparaci√≥n de medias entre dos poblaciones independientes, espec√≠ficamente la prueba t de Student, dado que se busca detectar cualquier diferencia, sin especificar cu√°l g√©nero tiene calificaciones mayores o menores

```python
# Hip√≥tesis 1

# Filtrar los datos
scores_xbox_one = games[games['platform'] == 'XOne']['user_score'].dropna()
scores_pc = games[games['platform'] == 'PC']['user_score'].dropna()

# Nivel de significancia
alpha = 0.05

# Prueba de igualdad de varianzas (Levene)
stat_levene, p_levene = st.levene(scores_xbox_one, scores_pc)

print(f"Estad√≠stico Levene: {stat_levene : .4f}")
print(f"P-valor Levene: {p_levene : .4f}")

# Decidir si las varianzas son iguales
if p_levene > alpha:
    print("Las varianzas son iguales (no se rechaza H0). Se puede usar la prueba t con varianzas iguales.")
else:
    print("Las varianzas no son iguales (se rechaza H0). Se recomienda usar la prueba t sin asumir varianzas iguales.")
```

```python
# Prueba t student 

# Filtrar los datos
scores_xbox_one = games[games['platform'] == 'XOne']['user_score'].dropna()
scores_pc = games[games['platform'] == 'PC']['user_score'].dropna()

# Nivel de Significancia
alpha =0.05

t_stat, p_value = st.ttest_ind(scores_xbox_one, scores_pc, equal_var=True)
print(f"Estad√≠stico t: {t_stat:.4f}")
print(f"P-valor: {p_value:.4f}")

# Interpretaci√≥n final
if p_value < alpha:
    print("Se rechaza la hip√≥tesis nula: hay una diferencia significativa en las calificaciones promedio entre Xbox One y PC.")
else:
    print("No se puede rechazar la hip√≥tesis nula: no hay una diferencia significativa en las calificaciones promedio entre Xbox One y PC.")
```

```python
# Hip√≥tesis 2

# Prueba de Levene, para Determinar si las Varianzas de los Dos Grupos son Iguales.

# Filtrar las calificaciones por g√©nero y eliminar Valores Ausentes
action_scores = games[games["genre"] == "Action"]["user_score"].dropna()
sports_scores = games[games["genre"] == "Sports"]["user_score"].dropna()

# Nivel de Significancia
alpha = 0.05

# Prueba de Levene para igualdad de varianzas
stat_levene, p_levene = st.levene(action_scores, sports_scores)

print(f"Estad√≠stico de Levene: {stat_levene : .4f}")
print(f"Valor p de Levene: {p_levene : .4f}")

# Decidir si las varianzas son iguales
if p_levene > alpha:
    print("Las varianzas son iguales (no se rechaza H0). Se puede usar la prueba t con varianzas iguales.")
else:
    print("Las varianzas no son iguales (se rechaza H0). Se recomienda usar la prueba t sin asumir varianzas iguales.")
```
